```python
import requests 
from bs4 import BeautifulSoup
```

```python
url = "https://eresearch.fidelity.com/eresearch/goto/markets_sectors/landing.jhtml"
base = "https://eresearch.fidelity.com"
response = requests.get(url)
if response.status_code == 200:
    pass
else:
    print("Error connecting to webpage")
page = BeautifulSoup(response.content,"lxml")
sectors = page.find_all("div",class_ = "heading")
for sector in sectors:
        all_links = sector.find("a").get("href")
        if "https" not in all_links:
            sector_url = base + all_links
            sector_response = requests.get(sector_url)
            if response.status_code == 200:
                pass
            else:
                print("Error connecting to webpage")
            response = requests.get(sector_url)
            page = BeautifulSoup(response.content, "lxml")
            ratios = page.find_all("div", class_="sec-fundamentals")
            titles= page.find_all("div")
            for title in titles:
                if title.find("h1") != None:
                    print(title.find("h1").get_text() + ":")
                    print()
                    break
            values = ratios[0].find_all("td")
            value_list = []
            for value in values:
                value_list.append(value.get_text().strip())
            value_list = value_list[:-1]


            key_list = []
            keys = ratios[0].find_all("th")
            for key in keys:
                key_list.append(key.get_text())

            key_list= key_list[2:]
            metrics = dict(zip(key_list,value_list))
            for key, value in metrics.items():
                print(key+":",value)
            print()
```
